136. Single Number

Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

 

Example 1:

Input: nums = [2,2,1]

Output: 1

### LeetCode 136 — Single Number
Topic: Bit Manipulation (XOR)
Difficulty: Easy

Approach:
Use XOR to cancel out duplicate numbers. XOR has two important properties:
1. a ^ a = 0 (a number XOR itself becomes zero)
2. a ^ 0 = a

Since every number appears twice except one, all duplicates cancel out,
and the remaining value is the single number.

Time Complexity: O(n)
Space Complexity: O(1)
"""


```python
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        res = 0
        for n in nums:
            res ^= n
        return res       
```
