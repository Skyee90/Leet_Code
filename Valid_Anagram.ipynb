## 242. Valid Anagram

Platform: LeetCode  |  Difficulty: Easy  |  Topic: String, Hash Table


📌 Problem Statement
Given two strings s and t, return true if t is an anagram of s, and false otherwise.

🔤 An anagram is a word formed by rearranging the letters of another word, using all original letters exactly once.

Example 1:
Input:  s = "anagram", t = "nagaram"
Output: True  ✅
Example 2:
Input:  s = "rat", t = "car"
Output: False ❌

💡 Intuition
Think of it like counting coins 🪙
Imagine s and t are two bags of letter-tiles.

You count every letter in bag s
You count every letter in bag t
If both bags have the exact same counts → they're anagrams ✅

```python
s = "anagram"        t = "nagaram"

a → 3                a → 3   ✅
n → 1                n → 1   ✅
g → 1                g → 1   ✅
r → 1                r → 1   ✅
m → 1                m → 1   ✅

All counts match → True ✅
```

```python
✅ Solution
pythonclass Solution:
    def isAnagram(self, s: str, t: str) -> bool:

        # ⚡ Early exit — different lengths can't be anagrams
        if len(s) != len(t):
            return False

        # 📦 Two empty dictionaries to store letter counts
        countS, countT = {}, {}

        # 🔢 Count every character in both strings simultaneously
        for i in range(len(s)):
            countS[s[i]] = 1 + countS.get(s[i], 0)
            countT[t[i]] = 1 + countT.get(t[i], 0)

        # 🔍 Compare counts — every letter in s must match t
        for c in countS:
            if countS[c] != countT.get(c, 0):
                return False

        return True
```

```python
🧹 Cleaner Alternative (Python One-liner)

pythonfrom collections import Counter

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        return Counter(s) == Counter(t)
```
