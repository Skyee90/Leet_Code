1. Two Sum
Platform: LeetCode  |  Difficulty: Easy  |  Topic: Array, Hash Table

📌 Problem Statement

Given an array of integers nums and an integer target, return the indices of the two numbers that add up to target.
You may assume exactly one solution exists, and you cannot use the same element twice.

Input:  nums = [2, 7, 11, 15],  target = 9

Output: [0, 1]

Reason: nums[0] + nums[1] = 2 + 7 = 9 ✅


💡 Intuition

Think of it like finding a missing puzzle piece 🧩
You're walking through the array one number at a time.
For each number, you ask:

"Have I already seen the number I need to pair with this one?"

That missing number is called the complement → complement = target - num
You keep a notebook (dictionary) of every number you've seen so far and where you saw it.
The moment you find a number whose complement is already in your notebook → you're done! ✅

nums = [2, 7, 11, 15],  target = 9


Step 1 → num=2,  complement = 9-2 = 7  → 7 not seen yet, store {2:0}

Step 2 → num=7,  complement = 9-7 = 2  → 2 is in notebook at index 0 🎯

Return [0, 1]

```python
✅ Solution
class Solution:
    def twoSum(self, nums, target):

        seen = {}                          # 📒 Our notebook: { number : index }

        for i, num in enumerate(nums):     # 🔁 Walk through each number with its index

            complement = target - num      # 🧩 What number do we need to complete the pair?

            if complement in seen:         # ✅ If we've seen it before
                return [seen[complement], i]   # 🎯 Return both indices

            seen[num] = i                  # 📝 Haven't found pair yet, save this number

        return None                        # No solution found (won't happen per problem rules)
```

### ⏱ Time Complexity: O(n)

| Step | Cost | Reason |
|-----|-----|-----|
| Loop through `nums` | O(n) | Visit each element once |
| Dictionary lookup | O(1) | Hash map lookup is constant time |
| **Total** | **O(n)** | Only one pass through the array |

💡 **Comparison**

- Brute force approach (nested loops): **O(n²)**
- Optimized approach with hash map: **O(n)**


### 💾 Space Complexity — O(n)

| What | Space | Why |
|-----|-----|-----|
| `seen` dictionary | O(n) | Worst case stores all n elements |
| **Total** | **O(n)** | Grows with input size |

💡 In the worst case (pair is at the very end), we store almost every number before finding the answer.

### ⚖️ Approach Comparison

| Approach | Time | Space | Notes |
|---------|------|------|------|
| Hash Map (this solution) | O(n) | O(n) | Fast, one pass ✅ |
| Brute Force (nested loops) | O(n²) | O(1) | Simple but slow ❌ |
| Sorting + Two Pointers | O(n log n) | O(1) | Doesn't return original indices ⚠️ |

### 🔑 Key Concepts to Remember

| Concept | What it means |
|--------|---------------|
| 📒 Hash Map | Dictionary that gives O(1) lookup — perfect for "have I seen this before?" |
| 🧩 Complement | `target - num` — the missing piece we're searching for |
| 🔁 `enumerate()` | Pythonic way to get both index and value in a loop |
| ⚡ One Pass | We solve it in a single loop — no going back |
