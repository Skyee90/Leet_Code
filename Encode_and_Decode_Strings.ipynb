# 271. Encode and Decode Strings

**Platform:** LeetCode &nbsp;|&nbsp; **Difficulty:** Medium &nbsp;|&nbsp; **Topic:** Array, String

---

## 📌 Problem Statement

Design an algorithm to **encode** a list of strings into a single string, and **decode** that single string back into the original list.

The encoded string is sent over a network — the decoder must perfectly reconstruct the original list.

    encode(["neet","code","love","you"])  →  "4#neet4#code4#love3#you"

    decode("4#neet4#code4#love3#you")    →  ["neet","code","love","you"]

---

## 💡 Intuition

Think of it like shipping labelled packages 📦

When you encode, you attach a **label** to each word saying how long it is:

    "neet"  →  "4#neet"      (4 letters, then #, then the word)
    "code"  →  "4#code"
    "love"  →  "4#love"
    "you"   →  "3#you"

    Full encoded string: "4#neet4#code4#love3#you"

When you decode, you read the label first to know **exactly how many characters to grab** — then jump to the next label. No guessing, no ambiguity. 🎯

Why not just use a separator like `","` or `" "`?

    ❌ Problem: what if a word itself contains that separator?
    ["hello,world", "hi"]  →  "hello,world,hi"  →  impossible to decode correctly!

    ✅ Our fix: length prefix tells us exactly where each word ends — no separator needed.

---

## ✅ Solution

    class Solution:

        def encode(self, strs: List[str]) -> str:

            res = ""
            for s in strs:                          # 🔁 Go through each word
                res += str(len(s)) + "#" + s        # 📦 Attach label: length + # + word
            return res


        def decode(self, s: str) -> List[str]:

            res, i = [], 0                          # 📋 Result list, pointer starts at 0

            while i < len(s):                       # 🔁 Keep going until end of string

                j = i
                while s[j] != "#":                  # 🔍 Scan forward to find the #
                    j += 1

                length = int(s[i:j])                # 🔢 Read the number before #
                res.append(s[j+1 : j+1+length])     # ✂️  Slice exactly that many characters
                i = j + 1 + length                  # ⏭️  Jump pointer to next label

            return res

---

## 🔍 Line-by-Line Breakdown

---

### 🔐 ENCODE

**Build the encoded string word by word**

    res += str(len(s)) + "#" + s

For each word `s`:
- `len(s)` → how many characters the word has
- `"#"` → separator between the number and the word
- `s` → the actual word

    s = "neet"  →  str(4) + "#" + "neet"  =  "4#neet"
    s = "you"   →  str(3) + "#" + "you"   =  "3#you"

    Final: "4#neet4#code4#love3#you"

---

### 🔓 DECODE

**Two pointers: `i` marks the start of each label, `j` scans to find `#`**

    encoded = "4#neet4#code4#love3#you"
               ^
               i starts here

---

**Step 1 — Find the `#`**

    j = i
    while s[j] != "#":
        j += 1

Move `j` forward until it hits `#`. Everything between `i` and `j` is the length number.

    i=0, scan forward:  s[0]='4', s[1]='#'  →  j stops at 1
    s[i:j] = s[0:1] = "4"

---

**Step 2 — Read the Length**

    length = int(s[i:j])

Convert the string number to an integer — this tells us how many characters to grab.

    s[0:1] = "4"  →  length = 4

---

**Step 3 — Slice the Word**

    res.append(s[j+1 : j+1+length])

- `j+1` → skip past the `#`
- `j+1+length` → grab exactly `length` characters

    j=1, length=4
    s[2 : 2+4] = s[2:6] = "neet" ✅

---

**Step 4 — Move the Pointer**

    i = j + 1 + length

Jump `i` to the start of the **next label** and repeat.

    i = 1 + 1 + 4 = 6   →  now pointing at "4#code..."

---

**Full Walkthrough**

    encoded = "4#neet4#code4#love3#you"

    Round 1: i=0,  j=1,  length=4  →  grab s[2:6]   = "neet",  i jumps to 6
    Round 2: i=6,  j=7,  length=4  →  grab s[8:12]  = "code",  i jumps to 12
    Round 3: i=12, j=13, length=4  →  grab s[14:18] = "love",  i jumps to 18
    Round 4: i=18, j=19, length=3  →  grab s[20:23] = "you",   i jumps to 23

    Result: ["neet", "code", "love", "you"] ✅

## ⏱️ Time Complexity — O(n)

| Method | Cost | Why |
|--------|------|-----|
| `encode` | O(n) | Visit each character once across all words |
| `decode` | O(n) | Scan through the encoded string once |
| **Total** | **O(n)** | n = total number of characters across all strings |

---

## 💾 Space Complexity — O(n)

| What | Space | Why |
|------|-------|-----|
| Encoded string | O(n) | Stores all characters + length labels |
| Decoded list | O(n) | Stores all original words |
| **Total** | **O(n)** | Proportional to total input size |

---

## 🔑 Key Concepts to Remember

| Concept | What it means |
|---------|---------------|
| 📦 **Length Prefix Encoding** | Attach length before each word so decoder knows exactly where it ends |
| 🔍 **Two Pointer Scan** | `i` = start of label, `j` = scans to find `#` |
| ✂️ **String Slicing** | `s[j+1 : j+1+length]` grabs exactly the right word |
| ⚠️ **Why not use a delimiter?** | Words may contain any character — length prefix is the only safe approach |
