# 347. Top K Frequent Elements

**Platform:** LeetCode &nbsp;|&nbsp; **Difficulty:** Medium &nbsp;|&nbsp; **Topic:** Array, Hash Table, Sorting

---

## 📌 Problem Statement

Given an integer array `nums` and an integer `k`, return the `k` most frequent elements.

    Input:  nums = [1,1,1,2,2,3],  k = 2
    Output: [1, 2]
    Reason: 1 appears 3 times, 2 appears 2 times → top 2 most frequent

    Input:  nums = [1],  k = 1
    Output: [1]

---

## 💡 Intuition

Think of it like a school leaderboard 🏆

Imagine a scoreboard with rows numbered 1, 2, 3 ... up to `n`.
Each row represents a **frequency** (how many times a number appeared).
Every number gets placed in the row matching its count.

    nums = [1,1,1,2,2,3]

    Row 1 → [3]        ← 3 appeared 1 time
    Row 2 → [2]        ← 2 appeared 2 times
    Row 3 → [1]        ← 1 appeared 3 times

To get the top `k` frequent → just **read from the bottom up** and collect until you have `k` numbers. 🎯

This clever trick is called **Bucket Sort** — and it avoids expensive sorting entirely!

---

## ✅ Solution

    class Solution:
        def topKFrequent(self, nums: List[int], k: int) -> List[int]:

            count = {}                              # 📊 Count how often each number appears
            freq = [[] for i in range(len(nums)+1)] # 🪣 Buckets: index = frequency

            for n in nums:                          # 🔢 Build the frequency count
                count[n] = 1 + count.get(n, 0)

            for n, c in count.items():              # 🪣 Place each number in its bucket
                freq[c].append(n)

            res = []
            for i in range(len(freq)-1, 0, -1):    # 🔽 Walk buckets from highest to lowest
                for n in freq[i]:
                    res.append(n)
                    if len(res) == k:               # 🎯 Stop as soon as we have k results
                        return res

---

## 🔍 Line-by-Line Breakdown

**Step 1 — Count Frequencies**

    count = {}

    for n in nums:
        count[n] = 1 + count.get(n, 0)

Walk through `nums` and count how many times each number appears.

    nums = [1, 1, 1, 2, 2, 3]

    n=1 → count = {1: 1}
    n=1 → count = {1: 2}
    n=1 → count = {1: 3}
    n=2 → count = {1: 3, 2: 1}
    n=2 → count = {1: 3, 2: 2}
    n=3 → count = {1: 3, 2: 2, 3: 1}

---

**Step 2 — Create the Buckets 🪣**

    freq = [[] for i in range(len(nums) + 1)]

Create a list of empty lists. The **index represents frequency**.

Why `len(nums) + 1`?
The maximum possible frequency of any number is `len(nums)` (if all elements are the same).
So we need slots from index `0` to index `len(nums)`.

    nums has 6 elements → freq has 7 slots (index 0 to 6)

    freq = [ [], [], [], [], [], [], [] ]
             0    1    2    3    4    5    6

---

**Step 3 — Fill the Buckets**

    for n, c in count.items():
        freq[c].append(n)

For each number `n` with count `c`, place `n` into bucket at index `c`.

    count = {1: 3, 2: 2, 3: 1}

    n=1, c=3 → freq[3].append(1)
    n=2, c=2 → freq[2].append(2)
    n=3, c=1 → freq[1].append(3)

    freq = [ [], [3], [2], [1], [], [], [] ]
             0    1    2    3   4   5   6

---

**Step 4 — Read From Highest Frequency Down 🔽**

    for i in range(len(freq)-1, 0, -1):

This loops from the **last index down to 1** (skipping 0 since frequency 0 is impossible).

    len(freq) = 7  →  loop: i = 6, 5, 4, 3, 2, 1

We go from the highest bucket downward because we want the **most frequent** elements first.

---

**Step 5 — Collect Until We Have k Results 🎯**

    for n in freq[i]:
        res.append(n)
        if len(res) == k:
            return res

Pick numbers from the current bucket, add to result.
The moment we have `k` numbers → return immediately. No need to go further.

    k = 2

    i=6 → freq[6] = []  → skip
    i=5 → freq[5] = []  → skip
    i=4 → freq[4] = []  → skip
    i=3 → freq[3] = [1] → res = [1], len(res)=1, not done yet
    i=2 → freq[2] = [2] → res = [1,2], len(res)=2 == k → return [1, 2] 🎯

---

## ⏱️ Time Complexity — O(n)

| Step | Cost | Why |
|------|------|-----|
| Count frequencies | O(n) | One pass through nums |
| Fill buckets | O(n) | One pass through count dictionary |
| Read buckets | O(n) | At most visit all n elements |
| **Total** | **O(n)** | Everything is linear! |

💡 A typical sorting approach would be O(n log n).
Bucket Sort lets us skip sorting entirely and stay at O(n) — that's the whole point of this approach!

---

## 💾 Space Complexity — O(n)

| What | Space | Why |
|------|-------|-----|
| `count` dictionary | O(n) | Stores up to n unique numbers |
| `freq` bucket list | O(n) | n+1 buckets total |
| `res` result list | O(k) | At most k elements |
| **Total** | **O(n)** | Scales with input size |

---

## ⚖️ Approach Comparison

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| **Bucket Sort (this solution)** | O(n) | O(n) | Fastest possible ✅ |
| Sort by frequency | O(n log n) | O(n) | Simpler code, slower |
| Min-Heap | O(n log k) | O(n) | Good when k is very small |

## 🔑 Key Concepts to Remember

| Concept | What it means |
|---------|---------------|
| 🪣 **Bucket Sort** | Use array index as the value — avoids comparison-based sorting |
| 📊 **Frequency Map** | Dictionary counting occurrences of each element |
| 🔽 **Reverse Traversal** | Reading high-index buckets first = highest frequency first |
| 🎯 **Early Exit** | Stop collecting the moment we hit exactly k results |
