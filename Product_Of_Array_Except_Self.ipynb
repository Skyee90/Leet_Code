# 238. Product of Array Except Self

**Platform:** LeetCode &nbsp;|&nbsp; **Difficulty:** Medium &nbsp;|&nbsp; **Topic:** Array, Prefix Sum

---

## 📌 Problem Statement

Given an integer array `nums`, return an array `answer` where `answer[i]` is the **product of all elements except** `nums[i]`.

You must solve it **without using division** and in **O(n)** time.

    Input:  nums = [1, 2, 3, 4]
    Output: [24, 12, 8, 6]

    Explanation:
    index 0 → 2 × 3 × 4 = 24   (everything except nums[0]=1)
    index 1 → 1 × 3 × 4 = 12   (everything except nums[1]=2)
    index 2 → 1 × 2 × 4 = 8    (everything except nums[2]=3)
    index 3 → 1 × 2 × 3 = 6    (everything except nums[3]=4)

---

## 💡 Intuition

Think of it like two people walking past a shelf 🚶‍♂️🚶‍♀️

Imagine the array as a row of boxes. For each box, you need the product of everything to its **left** AND everything to its **right**.

    nums = [1, 2, 3, 4]

    LEFT products  (prefix):   [1,  1,  2,  6 ]
                                ↑   ↑   ↑   ↑
                               nothing  1  1×2  1×2×3

    RIGHT products (postfix):  [24, 12,  4,  1 ]
                                ↑    ↑   ↑   ↑
                            2×3×4  3×4   4  nothing

    Answer = prefix × postfix: [1×24, 1×12, 2×4, 6×1] = [24, 12, 8, 6] ✅

Two passes. No division. No extra array needed. 🎯

---

## ✅ Solution

    class Solution:
        def productExceptSelf(self, nums: List[int]) -> List[int]:

            res = [1] * len(nums)    
                    
                       # 📋 Start    with all 1s

            prefix = 1
            for i in range(len(nums)): 
            
                  # ➡️  Left to right pass — fill    prefix products

                res[i] = prefix
                prefix *= nums[i]

            postfix = 1
            for i in range(len(nums)-1, -1, -1): 
            
                        # ⬅️  Right to left pass — multiply     postfix

                res[i] *= postfix
                postfix *= nums[i]

            return res

## 🔍 Line-by-Line Breakdown

---

**Step 1 — Initialize Result Array**

    res = [1] * len(nums)

Create a result array filled with 1s — same length as `nums`.
We'll build the answer directly in this array across two passes.

    nums = [1, 2, 3, 4]
    res  = [1, 1, 1, 1]   ← starting point

---

**Step 2 — Left to Right Pass (Prefix) ➡️**

    prefix = 1
    for i in range(len(nums)):
        res[i] = prefix
        prefix *= nums[i]

Walk left to right. At each position `i`, store the product of everything **to the left** of `i`.
Then update `prefix` to include `nums[i]` for the next round.

    i=0: res[0] = 1        prefix = 1 × 1 = 1
    i=1: res[1] = 1        prefix = 1 × 2 = 2
    i=2: res[2] = 2        prefix = 2 × 3 = 6
    i=3: res[3] = 6        prefix = 6 × 4 = 24

    res after prefix pass = [1, 1, 2, 6]

Notice: `res[0] = 1` because there's **nothing to the left** of index 0.

---

**Step 3 — Right to Left Pass (Postfix) ⬅️**

    postfix = 1
    for i in range(len(nums)-1, -1, -1):
        res[i] *= postfix
        postfix *= nums[i]

Walk right to left. At each position `i`, **multiply** the existing prefix value by the product of everything **to the right** of `i`.
Then update `postfix` to include `nums[i]` for the next round.

    i=3: res[3] = 6  × 1  = 6     postfix = 1 × 4 = 4
    i=2: res[2] = 2  × 4  = 8     postfix = 4 × 3 = 12
    i=1: res[1] = 1  × 12 = 12    postfix = 12 × 2 = 24
    i=0: res[0] = 1  × 24 = 24    postfix = 24 × 1 = 24

    res after postfix pass = [24, 12, 8, 6] ✅

---

**Full Picture Side by Side**

    nums    =  [  1,   2,   3,   4  ]

    prefix  =  [  1,   1,   2,   6  ]   ← product of everything LEFT
    postfix =  [ 24,  12,   4,   1  ]   ← product of everything RIGHT

    answer  =  [ 24,  12,   8,   6  ]   ← prefix × postfix at each index

---

## ⏱️ Time Complexity — O(n)

| Step | Cost | Why |
|------|------|-----|
| Prefix pass | O(n) | One left-to-right loop |
| Postfix pass | O(n) | One right-to-left loop |
| **Total** | **O(n)** | Two linear passes = still O(n) |

💡 The brute force approach (multiply everything for each index) is O(n²).
Two passes cut that down to O(n) with zero division.

---

## 💾 Space Complexity — O(1)

| What | Space | Why |
|------|-------|-----|
| `res` output array | O(n) | Required for output — doesn't count as extra space |
| `prefix` / `postfix` | O(1) | Just two variables |
| **Total** | **O(1)** | No extra data structures used |

💡 The problem says the output array doesn't count toward space complexity.
That's why this is considered O(1) extra space — we reuse `res` for both passes.

## 🔑 Key Concepts to Remember

| Concept | What it means |
|---------|---------------|
| ➡️ **Prefix Product** | Product of all elements to the LEFT of index i |
| ⬅️ **Postfix Product** | Product of all elements to the RIGHT of index i |
| 🔄 **Two Pass Trick** | First pass stores prefix, second pass multiplies in postfix |
| ♻️ **In-place reuse** | We use the output array itself — no separate prefix/postfix arrays needed |
