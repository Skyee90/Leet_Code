217. Contains Duplicate

Platform: LeetCode  |  Difficulty: Easy  |  Topic: Arrays, Hashing


📌 Problem Statement
Given an integer array nums, return true if any value appears at least twice, and return false if every element is distinct.


Example 1:
Input:  nums = [1, 2, 3, 1]
Output: True

Example 2:
Input:  nums = [1, 2, 3, 4]
Output: False

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(nums) != len(set(nums))
```

🧠 Intuition
The core idea is simple:

A set in Python stores only unique elements. So if we convert nums into a set, any duplicate values will be automatically removed.
If the length of the original list differs from the length of the set — it means some elements were removed, i.e., duplicates existed.

🔑 Key Concepts to Remember

* Set → Stores only unique values. Duplicate insertion is silently ignored.

* len() on list vs set → A fast way to detect if any value was deduplicated.

* Hash Set under the hood → Python sets use hashing, giving O(1) average insert/ lookup.
